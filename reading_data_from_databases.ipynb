# Reading Data from Databases

```python
import pandas as pd
import sqlite3
```

```python
# Connecting to our database

conn = sqlite3.connect("data/chinook.db")

# Using the cursor function, we can now execute SQL commands
cur = conn.cursor()

# Selecting employees from the database

cur = conn.execute("SELECT * FROM employees")

# The fetch gets you the results of your last query
result = cur.fetchall()
result
```

```python
# Creating a dataframe from the resultss
df = pd.DataFrame(result)

df
```

```python
# Another good way to read sql files is by read_sql

df = pd.read_sql("SELECT * FROM employees",conn,
    index_col="EmployeeId",
    parse_dates=["BirthDate","HireDate"])

df
```
