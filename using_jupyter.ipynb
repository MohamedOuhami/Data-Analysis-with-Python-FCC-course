Importing the needed libraries fro visualization ( Numpy and Matplotlib)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = np.linspace(0,10,500)
y=np.cumsum(np.random.randn(500,6),0)
```

```python
plt.figure(figsize=(12,7))
plt.plot(x,y)
```

You can see clearly the graph It gave you based on the data It got. That is why It is such a beautiful tool
